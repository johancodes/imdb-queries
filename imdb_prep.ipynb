```python
merged_df = pd.merge(df_basics, df_ratings, on='tconst')

merged_df
```

```python
# merged_df.info() # there are no NaN, and I use code below anyway to prove it.
# number of rows have not changed, hence no NaN.

merged_df = merged_df.dropna()

merged_df
```

```python
# let's keep the dataset "clean" by removing adult titles
# first we work out how many adult titles there are

merged_df['isAdult'].value_counts()
```

```python
# drop adult films

merged_df = merged_df[ merged_df['isAdult']==0 ]
```

```python
highest_rating = merged_df[ merged_df['averageRating'] >= 9 ]

highest_rating = highest_rating.sort_values( 'averageRating', ascending=False)

highest_rating
```

```python
# drop isAdult column, now that all films are not adult

merged_df = merged_df.drop(columns=['isAdult'], axis=1)
```

```python
# inspect new csv as a pd dataframe

df_imdb20mar23 = pd.read_csv('imdb20mar23.csv')

df_imdb20mar23.iloc[:5 , :]
```

```python
# inspect datatypes 

df_imdb20mar23.dtypes
```

```python
# startYear is an object (meaning it contains more than one kind of data) and this will be
# a problem later. We want to convert all startYear values into integers. First thing
# we need to do is to deal with the \N values. so we first replace all \N values with 0. 
# We replace ALL instances of \N with 0, in whichever column it appears
# !!! note the r before '\N' 

df_imdb20mar23 = df_imdb20mar23.replace(r'\N', 0)
```

```python
df_imdb20mar23

# all instances of \N has been replaced with 0
```

```python
df_imdb20mar23.dtypes

# startYear is still an object
```

```python
# now we convert all startYear values into integers

df_imdb20mar23['startYear'] = df_imdb20mar23['startYear'].astype('int64')
```

```python
df_imdb20mar23.dtypes

# startYear is now an integer!
```

```python
# create a new csv from this dataframe

df_imdb20mar23.to_csv('imdb20mar23_b.csv', index=False)
```

```python
# use 'imdb20mar23_b.csv'

df_imdb20mar23_b = pd.read_csv('imdb20mar23_b.csv')
```

```python
# my query filters

mask_1 = df_imdb20mar23["genres"].str.contains("Drama")
mask_2 = df_imdb20mar23['averageRating'] >= 7
mask_3 = df_imdb20mar23['startYear'] >= 2020 
mask_4 = df_imdb20mar23['titleType'].str.contains("movie")

# use query filters 

ratings_and_genre = df_imdb20mar23_b[ mask_1 & mask_2 & mask_3 & mask_4 ]

# sort by rating

ratings_and_genre.sort_values('averageRating', ascending=False)

```
